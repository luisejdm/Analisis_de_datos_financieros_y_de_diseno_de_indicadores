```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Actividad 1

```python
data = pd.read_csv('customer_data.csv')
data.head()
```

```python
data.groupby('Segment')['NPS_Score'].mean().plot(kind='bar', color='skyblue')
plt.title('Score Promedio por Segmento', fontsize=17)
plt.xlabel('Segmento')
plt.ylabel('Score Promedio')
plt.show()
```

```python
data.groupby('Segment')['Expected_Lifespan_Months'].mean().plot(kind='bar', color='skyblue')
plt.title('Promedio de Expected_Lifespan_Months', fontsize=17)
plt.xlabel('Segmento')
plt.ylabel('Expected_Lifespan_Months')
plt.show()
```

### Conclusión

En promedio, el segmento "Senior" en cuanto a proporción, es casi igual de bajo en Lifespan Esperado y NPS comparado a los otros segmentos.

Dado que se quiere mejorar el LTV esperado, habría que poner atención en el segmento de seniors porque:

- En promedio tienen calificación de 3 comparado a los otros segmentos que tienen aproximadamente 6.
- En promedio tienen un lifespan de 8 meses comparado a los otros que tienen aproximadamente 14.

Por lo tanto, mejorar el servicio, y estrategias que impacten al sector senior.

## Actividad 2

```python
data['ltv/cac'] = data['Expected_LTV_in_Two_Years'] / data['CAC']
data.head()
```

```python
data3 = data.groupby(['Acquisition_Month', 'Segment'])['CAC'].mean().reset_index()
sns.lineplot(x='Acquisition_Month', y='CAC', data=data3, hue='Segment' , palette='Set3')
plt.title('Distribución de la Factura Total por Día')
plt.show()
```

```python
data2 = data.groupby(['Acquisition_Month', 'Segment'])['Expected_LTV_in_Two_Years'].mean().reset_index()
sns.lineplot(x='Acquisition_Month', y='Expected_LTV_in_Two_Years', data=data2, hue='Segment' , palette='Set3')
plt.title('Expected_LTV_in_Two_Years por segmento')
plt.show()
```

```python
data4 = data.groupby(['Acquisition_Month', 'Segment'])['ltv/cac'].mean().reset_index()
sns.lineplot(x='Acquisition_Month', y='ltv/cac', data=data4, hue='Segment' , palette='Set3')
plt.title('LTV/CAC promedio por segmento')
plt.show()
```

```python
data4.groupby('Segment')['ltv/cac'].mean()
```

### Conclusión

- El CAC es igual para los 3 segmentos.
- El LTV esperado por segmento nuevamente es mas bajo en el segmento de Seniors.
- A pesar de que el LTV/CAC de los Senior se muestran rentables, su promedio está en el 1.82, mientras que los otros segmentos están alrededor de 3.

### Estrategia Final

- No gastar la misma cantidad en publicidad para los 3 segmentos de clientes, sino que gastar mas en el segmento Senior, y tratar de subir su LTV/CAC el 40% que le falta para ser igual de rentable que los otros 2 segmentos.
